```python
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv('C:/Users/nikhi/Desktop/blinkit data analysis and dashboard/BlinkIT Grocery Data.csv')
```

```python
df.sample(5)
# I found so many Nan values in the Item Weight column, It should be filled before data analysis.
# Data cleaning is required in the Item Fat Content column, eg. - reg, low fat, LF.
# Found two unnamed columns, should be removed.
```

```python
df.drop(['Unnamed: 12', 'Unnamed: 13'], axis=1, inplace=True)
```

```python
df["Item Weight"] = df["Item Weight"].replace("", np.nan)
```

```python
df["Item Fat Content"] = df["Item Fat Content"].replace({
    "LF": "Low Fat",
    "low fat": "Low Fat",
    "reg": "Regular"
})
```

```python
# Creating a mapping of Item Identifier of Item Weight using mean in case multiple weights exist per item
item_weight_map = df.groupby("Item Identifier")["Item Weight"].mean()
```

```python
item_weight_map
```

```python
# Filling missing values in Item Weight using the mapping
df["Item Weight"] = df["Item Weight"].fillna(df["Item Identifier"].map(item_weight_map))
```

```python
df.to_csv("blinkit_data", index = False)
```
